```python
#use Elbow method to determine the value of k
from sklearn.cluster import KMeans
inertia_list = []
for k in range(2,16):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(movements)
    inertia_list.append(kmeans.inertia_)

#plot the data
plt.plot(range(2,16), inertia_list)
plt.title("Elbow curve")
plt.xlabel("No of clusters")
plt.ylabel("Distortion")
plt.grid(True)
plt.show()
```

```python
c=5
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn.cluster import KMeans
normalizer = Normalizer()
kmeans = KMeans(n_clusters = c, max_iter = 1000, random_state=1)
# Make a pipeline chaining normalizer and kmeans
pipeline = make_pipeline(normalizer,kmeans)
# Fit pipeline to daily stock movements
pipeline.fit(movements)
predictions = pipeline.predict(movements)
labels = kmeans.labels_
labels
```

```python
#Calculate returns and create a data frame
returns = df.pct_change().mean()*252
returns = pd.DataFrame(returns)
returns.columns = ['returns']
length = len(companies_dict)
#Calculate the volatility
returns['volatility'] = df.pct_change().std()*np.sqrt(252)

data = returns
data = data.iloc[0:length,:]
data
```

```python
plt.figure(figsize = (20,8)) 
ax1 = plt.subplot(1,2,1)
plt.title('Returns',fontsize = 20)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 20)
plt.xlabel('Date',fontsize = 20)
plt.ylabel('Returns',fontsize = 20)
plt.plot(data['returns'].values) 
plt.subplot(1,2,2,sharey = ax1)
plt.title('Volatility',fontsize = 20)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 20)
plt.xlabel('Date',fontsize = 20)
plt.ylabel('Volatility',fontsize = 20)
plt.plot(data['volatility'].values)
```

Normalize the data

```python
from sklearn.preprocessing import StandardScaler

#Prepare the scaler
scale = StandardScaler().fit(data)

#Fit the scaler
scaled_data = pd.DataFrame(scale.fit_transform(data),columns = data.columns, index = data.index)
X = scaled_data
X
```

# Elbow curve to determine the value of k

```python
from sklearn.cluster import KMeans
data_norm = X[['volatility', 'returns']].values
inertia_list = []
import time
begin = time.time()
for k in range(2,24):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia_list.append(kmeans.inertia_)
end = time.time()
time_taken_elbow = end-begin
#plot the data
plt.plot(range(2,24), inertia_list)
plt.title("Elbow curve")
plt.xlabel("No of clusters")
plt.ylabel("Inertia")
plt.grid(True)
plt.show()
```

```python
c=6
import time
begin = time.time()
kmeans = KMeans(n_clusters=c)
kmeans.fit(X)
end = time.time()
labels = kmeans.labels_
time_taken_kmean = end-begin
labels
```

```python
#Fit the model
clusters = 5
begin = time.time()
hc = AgglomerativeClustering(n_clusters= clusters, affinity='euclidean', linkage='ward')
labels = hc.fit_predict(X)
end = time.time()
time_taken_hc = end-begin
#Plot the results
fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(111)
scatter = ax.scatter(X.iloc[:,0], X.iloc[:,1], c=labels, cmap='rainbow')
ax.set_title('Hierarchical Clustering Results')
ax.set_xlabel('Mean Return')
ax.set_ylabel('Volatility')
plt.colorbar(scatter)
plt.show()
```

```python
hc_validation = metric.silhouette_score(X, hc.fit_predict(X), metric='euclidean')
print("Hierarchical Clustering", metric.silhouette_score(X, hc.fit_predict(X), metric='euclidean'))
```
